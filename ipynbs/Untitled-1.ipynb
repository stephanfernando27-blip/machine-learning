```python
import pandas as pd
sleephealth_df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
sleephealth_df.info()
```

```python
sleephealth_df.head()
```

```python
import matplotlib.pyplot as plt

# Only keep numeric columns
numeric_df = sleephealth_df.select_dtypes(include='number')

fig, ax = plt.subplots(len(numeric_df.columns), 1, figsize=(7, 18), dpi=96)

for i, col in enumerate(numeric_df.columns):
    ax[i].boxplot(numeric_df[col].dropna(), vert=False)  # dropna() avoids NaN issues
    ax[i].set_title(col)

plt.tight_layout()
plt.show()
```
